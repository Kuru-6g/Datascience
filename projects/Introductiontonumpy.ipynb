**Dot Product Example(Nut Butter Sales**

```python
np.random.seed(0)
#No of jars sold
sales_amount = np.random.randint(20,size=(5,3))
sales_amount
```

```python
#Create a weekly salres Dataframe
weekly_sales = pd.DataFrame(sales_amount,index=["Mon","Tue","Thurs","Fri","Sat"],
                            columns=["AlmondButter","CashewButter","PeanutButter"])
weekly_sales
```

```python
#Create a price array
prices = np.array([10,18,13])
prices
```

```python
prices.shape
```

```python
#Create biutter prices dataframe
butter_prices = pd.DataFrame(prices.reshape(1,3),index=["Price"],columns=["AlmondButter","CashewButter","PeanutButter"])
butter_prices
```

```python
prices.shape
```

```python
sales_amount.shape
```

```python
total_sales = prices.dot(sales_amount)
total_sales
```

```python
#Let's Transpose
total_sales = prices.dot(sales_amount.T)
total_sales
```

```python
weekly_sales["Total Sales"] = total_sales
```

```python
weekly_sales
```

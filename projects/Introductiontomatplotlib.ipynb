**Plotting Data From pandas**

```python
import pandas as pd
car_sales = pd.read_csv("/content/drive/MyDrive/car-sales.csv")
car_sales
```

```python
#Remove dollars and comma from carsales dataframe
car_sales["Price"] = car_sales["Price"].str.replace('[$\,\.]','',regex=True)
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
#Add ndate column to carsales
car_sales["SalesDate"] = pd.date_range("1/1/2024",periods=len(car_sales))
car_sales
```

```python
#LOet's add another column
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x="SalesDate",y="Total Sales")
```

scatter

```python
#Change
car_sales.plot(x="Odometer (KM)",y="Price",kind="scatter");
```

Bar plot

```python
#Convert price to integer

car_sales.plot(x="Make",y="Odometer (KM)",kind="bar");
```

Histogram

```python
car_sales["Odometer (KM)"].hist()
```

**Plotting from pandas using OO Method**

```python
import pandas as pd
heart_disease = pd.read_csv("/content/drive/MyDrive/heart-disease.csv")
heart_disease
```

```python
#Filter out patients above 50
above_50 = heart_disease[heart_disease["age"] > 50]
above_50
```

```python
above_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             figsize=(10,6)
             )
```

```python
#Matplotlib workflow
#import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

#Create some data
#More data
# x = [1,2,3,4]
# y = [11,22,33,44]

#OOps method, Creating figiure and axes
fig , ax = plt.subplots(figsize=(5,5))

#plot data
scatter = ax.scatter(above_50["age"],above_50["chol"],c=above_50["target"])

#Customizing plot
ax.set(title="Heart Disease Analysis", xlabel="age", ylabel="Cholestrol")

#set legend
ax.legend(*scatter.legend_elements(),title="Target")

# #Save figure
# fig.savefig("/content/sample_data/heart_disease")

fig.suptitle("Heart Disease Analysis",fontweight="bold")
```

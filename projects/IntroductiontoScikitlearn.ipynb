scikitlearn workflow

```````````````python
#getting the data ready
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```````````````

```````````````python
#creates features and labels
X = heart_disease.drop("target",axis=1)
y = heart_disease["target"]
```````````````

```````````````python
clf.score(X_test,y_test)
```````````````

```````````````python
#metrics
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
print(classification_report(y_test,y_preds))
```````````````

```````````````python
#improving a model
np.random.seed(42)
for i in range(10,100,10):
  print(f"Trying model with {i} estimators....")
  clf = RandomForestClassifier(n_estimators=i).fit(X_train,y_train)
  print(f"Model accuracy on test set: {clf.score(X_test,y_test) * 100}% ")
```````````````

**What if missing values**

```````````````python
import pandas as pd
```````````````

```````````````python
car_sales_missing = pd.read_csv("/content/drive/MyDrive/car-sales-extended-missing-data.csv")
car_sales_missing
```````````````

```````````````python
car_sales_missing.isna().sum()
```````````````

```````````````python
#dropna
car_sales_missing.dropna(subset=["Price"],inplace=True)
```````````````

```````````````python
car_sales_missing.isna().sum()
```````````````

```````````````python
car_sales_missing["Doors"].value_counts()
```````````````

```````````````python
#Split into x and y
X = car_sales_missing.drop("Price",axis=1)
y = car_sales_missing["Price"]

#split into train and test splits
import numpy as np
np.random.seed(42)
from sklearn.model_selection import train_test_split
X_train,y_train,X_test,y_test = train_test_split(X,y,test_size=0.2)
```````````````

```````````````python
#check missing values
X.isna().sum() , y.isna().sum()
```````````````

```````````````python
#filling missing values with scikitlearn
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

#filling categorical columns as "missing" and numers as mean
cat_imputer = SimpleImputer(strategy="constant", fill_value="missing")
door_imputer = SimpleImputer(strategy="constant", fill_value=4)
num_imputer = SimpleImputer(strategy="mean")

#define columns
cat_features = ["Make", "Colour"]
door_feature = ["Doors"]
num_feature = ["Odometer (KM)"]

#create an imputer that fills missing values
imputer = ColumnTransformer([("cat_imputer",cat_imputer,cat_features),
                   ("door_imputer",door_imputer,door_feature),
                   ("num_imputer",num_imputer,num_feature )])


#fill trainvalues
imputer.fit_transform(X)
```````````````
